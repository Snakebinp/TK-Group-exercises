## 马尔萨斯模型

```python
import numpy as np
from scipy import optimize
import matplotlib.pyplot as plt

x0=3.9

def f1(t,r):
    return x0*np.exp(r*t)

x=np.array([3.9,5.3,7.2,9.6,12.9,17.1,23.2,31.4,38.6,50.2,62.9,76.0,92.0,106.5,123.2,131.7,150.7,179.3,204.0,226.5,251.4,281.4
])
t=np.arange(0,220,10)
fita,fitb=optimize.curve_fit(f1,t,x,maxfev=1000)  #optimize函数为scipy用于拟合自定义函数曲线的方法
```

```python
plt.scatter(t+1790, x, marker='o',label='real')
plt.plot(t+1790, f1(t,fita[0]),color='red',label='curve_fit')

plt.xlabel('Time')
plt.ylabel('Population')
```

```python
print('2010年人口预测数量为：',np.round(f1(220,fita[0]),2),'百万')
```

## 逻辑斯谛方程

```python
def f2(t,r,K):
    return K/(1+(K/x0-1)*np.exp(-r*t))

x=np.array([3.9,5.3,7.2,9.6,12.9,17.1,23.2,31.4,38.6,50.2,62.9,76.0,92.0,106.5,123.2,131.7,150.7,179.3,204.0,226.5,251.4,281.4
])
t=np.arange(0,220,10)
fita,fitb=optimize.curve_fit(f2,t,[],maxfev=1000)  #optimize函数为scipy用于拟合自定义函数曲线的方法
```

```python
plt.scatter(t+1790, x, marker='o',label='real')
plt.plot(t+1790, f2(t,fita[0],fita[1]),color='red',label='curve_fit')

plt.xlabel('Time')
plt.ylabel('Population')
```

```python
print('2010年人口预测数量为：',np.round(f2(220,fita[0],fita[1]),2),'百万')
```
